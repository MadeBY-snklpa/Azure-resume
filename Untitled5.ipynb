```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.tsa.statespace.sarimax import SARIMAX
from sklearn.model_selection import train_test_split




import statsmodels.api as sm
import matplotlib.pyplot as plt
# Step 1: Load the CSV file into a pandas DataFrame
data = pd.read_csv('/content/Alcohol_Sales.csv')  # Replace 'sales_data.csv' with your file name

# Step 2: Preprocess the data and set the 'date' column as the index
data['DATE'] = pd.to_datetime(data['DATE'])
data.set_index('DATE', inplace=True)

# Step 3: Optionally resample the data to handle irregular time intervals
#data = data.resample('D').sum()  # Resample to daily if data is not already in daily frequency

# Step 4: Split the data into training and testing sets
train_size = int(len(data) *1)
train, test = data[:train_size], data[train_size:]

# Step 5: Create and fit the SARIMA model
order = (1, 1, 1)            # (p, d, q) - Non-seasonal components
seasonal_order = (1, 1, 1, 12)  # (P, D, Q, S) - Seasonal components with a seasonal period of 12 (assuming monthly data)

model = SARIMAX(train, order=order, seasonal_order=seasonal_order)
results = model.fit()

# Step 6: Make predictions for the next 6 months
future_periods = 60
forecast = results.get_forecast(steps=future_periods)

# Step 7: Extract the predicted values and index from the forecast
y_pred = forecast.predicted_mean
forecast_index = pd.date_range(data.index[-1], periods=future_periods, freq='M')

```

```python
print(results.summary())
```

```python
plt.figure(figsize=(30, 6))
plt.plot(data.index, data['S4248SM144NCEN'], label='Actual')
plt.plot(forecast_index, y_pred, color='red', label='Predicted')
plt.xlabel('Date')
plt.ylabel('Sales')
plt.title('Actual vs. Predicted Sales using SARIMA')
plt.legend()
plt.show()
```
